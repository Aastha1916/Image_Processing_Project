```python
from skimage import io, color
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
from skimage.util import invert
```

```python
class Image_Editor():
    def __init__(self,np_image=None):
        self.np_image = np_image
        
    def Load_Image(self):
        print("Enter you Image file in the proper format(i.e file_name.extension):\n")
        self.np_image = io.imread(input()) 
        print(self.np_image.shape)
        print("IMAGE SUCCESSFULLY LOADED!!\n")
        fig, ax = plt.subplots()
        plt.imshow(self.np_image)
        plt.show()
        
        
    def Edit_Image(self):
        while True:
            print("1. Horizontal Flip\n2. Vertical Flip\n3. Invert Colors\n4. Convert to Grayscale\n5. Convert to B&W\n6. Return")
            choice = int(input("Enter your choice:\n"))
            if choice == 1:
                np_image_flip1 = self.np_image[::-1, :] #horizontal flip
                fig, ax = plt.subplots()
                plt.imshow(np_image_flip1)
                plt.show()
            elif choice == 2:
                np_image_flip2 = self.np_image[:, ::-1] #vertical flip
                fig, ax = plt.subplots()
                plt.imshow(np_image_flip2)
                plt.show()
            elif choice == 3:
                np_image_invert1 = invert((self.np_image)[:,:,:3])
                fig, ax = plt.subplots()
                plt.imshow(np_image_invert1)
                plt.show()
            elif choice == 4:
                gray_image = rgb2gray((self.np_image)[:,:,:3])
                fig, ax = plt.subplots()
                plt.imshow(gray_image, cmap='gray')
                plt.show()
            elif choice == 5:
                pass
            elif choice == 6:
                break
            else:
                print("INVALID CHOICE!!\n")  
                
    def Save_Image(self):
        io.imsave(fname="C:\\Users\\aastha agarwal\\Documents\\data_science_training\\Projects\\Untitled Folder\\saved_files\\save.png", arr=self.np_image)
        print("IMAGE SAVED!!\n")
        
            
    
    
```

```python
#Driver Code
print("_______________Image manipulation_______________ \n")
Aastha = Image_Editor()
while True:
    print("1. Load Image\n2. Edit Image\n3. Save Image\n4. Quit\n")
    choice = int(input("Enter your choice:\n"))
    if choice == 1:
        Aastha.Load_Image()
    elif choice == 2:
        Aastha.Edit_Image()
    elif choice == 3:
        Aastha.Save_Image()
    elif choice == 4:
        print("THANK YOU!!\nSEE YOU AGAIN.")
        break
    else:
        print("INVALID CHOICE!!\n")    
        
    
```
