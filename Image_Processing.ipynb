## Required Libraries

````python
from skimage import io
import matplotlib.pyplot as plt
import numpy as np

````

## Image_Editor class containing all required functions to perform following features:



````python
#Image_Editor class containing all required functions to perform following features:
class Image_Editor():
    def __init__(self,np_image=None, np_image_greyscale=None): #Constructor to initialize all required variables
        self.np_image = np_image #Variable to take input of images
      
    #Load Image code    
    def Load_Image(self):# To load images by selecting the appropriate input
        print("Select the Image file from the following availbale dataset:\n 1. Flower.jpg\n 2. Sunflower.jpg\n 3. Tiger.jpg\n")
        self.np_image = io.imread(input()) 
        print(self.np_image.shape)
        print("IMAGE SUCCESSFULLY LOADED!!\n")
        plt.imshow(self.np_image)
        plt.show()
        
     #Edit menu:   
    def Edit_Image(self):
        
        # Following features can be performed in the edit option:
        
        while True:
            
            print("1. Horizontal Flip\n2. Vertical Flip\n3. Invert Colors\n4. Convert to Grayscale\n5. Convert to B&W\n6. Return")
            choice = int(input("Enter your choice:\n"))
            
            if choice == 1:#horizontal flip
                np_image_flipped_horizontal = self.np_image[::-1, :] 
                print("IMAGE SUCCESSFULLY HORIZONTALLY FLIPPED!!\n")
                plt.imshow(np_image_flipped_horizontal)
                plt.show()
                
            elif choice == 2:#vertical flip
                np_image_flipped_vertical = self.np_image[:, ::-1] 
                print("IMAGE SUCCESSFULLY VERTICALLY FLIPPED!!\n")
                plt.imshow(np_image_flipped_vertical)
                plt.show()
                
            elif choice == 3: #inverting images
                np_image_inverted = self.np_image[:, :, ::-1]  
                print("IMAGE SUCCESSFULLY INVERTED!!\n")
                plt.imshow(np_image_inverted)
                plt.show()
                
            elif choice == 4: #Greyscale images
                np_image_greyscale = np.mean(self.np_image, axis=-1)
                #fig, ax = plt.subplots()
                print("IMAGE SUCCESSFULLY CONVERTED TO GREYSCALE!!\n")
                plt.imshow(np_image_greyscale, cmap='Greys')
                plt.show()
                
            elif choice == 5: #Black and white image 
                np_image_greyscale = np.mean(self.np_image, axis=-1) #First convert the colored image to greyscale
                np_image_blackandwhite = np.copy(np_image_greyscale)
                np_image_blackandwhite[np_image_blackandwhite < 128] = 0
                np_image_blackandwhite[np_image_blackandwhite >= 128] = 255
                print("IMAGE SUCCESSFULLY CONVERTED TO BLACK AND WHITE!!\n")
                plt.imshow(np_image_blackandwhite , cmap='Greys')
                plt.show()
                
            elif choice == 6:#Exit from the edit menu
                print("EDIT MENU TERMINATED!!\n")
                break
                
            else:#Try again with correct input
                print("INVALID CHOICE!!\n")  
                
    # Save image code      
    # These images are being saved in a particular directory and its path is mentioned her.
    def Save_Image(self):
        io.imsave(fname="C:\\Users\\aastha agarwal\\Documents\\data_science_training\\Projects\\Image_Processing_Project\\saved_files\\save.png", arr=self.np_image)
        print("IMAGE SAVED!!\n")
        
            
    
    
````

**NOTE:
* This Program runs for JPG Images only as others may have 4D.
* Keep the input images and Code in the Same Folder.
* You can load only Those Images which are present in the dataset.

## Driver Code

````python
#Driver Code
print("_______________Image manipulation_______________ \n")
#Class object is created
Aastha = Image_Editor()

#Main menu

while True:
    print("1. Load Image\n2. Edit Image\n3. Save Image\n4. Quit\n")
    choice = int(input("Enter your choice:\n"))
    # Input 1 to load images
    if choice == 1: 
        Aastha.Load_Image()
    #Input 2 to edit your loaded image
    elif choice == 2:
        Aastha.Edit_Image()
    #Input 3 to save your loaded image
    elif choice == 3:
        Aastha.Save_Image()
    #Input 4 to Exit
    elif choice == 4:
        print("THANK YOU!!\nSEE YOU AGAIN.")
        break
    #You have given some wrong input, Try again with appropriate input
    else:
        print("INVALID CHOICE!!\n")    
        
    
````
